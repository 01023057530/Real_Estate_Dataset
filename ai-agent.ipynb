```python
from autogen import  UserProxyAgent, AssistantAgent, GroupChat, GroupChatManager
```

# Agent Definitions

```python
# User Proxy Agent (acts as the frontend)
user_agent = UserProxyAgent(
    name = "UserAgent",
    system_message = "You help users book meeting rooms by collecting required details.",
    human_input_mode = "ALWAYS",
)
```

```python
# Availability Agent
availability_agent = AssistantAgent(
    name = "AvailabilityAgent",
    system_message = "You check room availability based on date, time, and duration. Return free rooms only.",
)
```

```python
# Feature Matching Agent
feature_agent = AssistantAgent(
    name = "FeatureAgent",
    system_message = "You match required room features (capacity, projector, video conferencing, etc.) with available rooms.",
)
```

```python
# Conflict Resolver Agent
conflict_agent = AssistantAgent(
    name = "ConflictAgent",
    system_message = "You resolve scheduling conflicts and suggest alternative time slots or rooms when needed.",
)
```

```python
# Booking Agent
booking_agent = AssistantAgent(
    name = "BookingAgent",
    system_message = "You finalize the room reservation and update the calendar/booking database.",
)
```

# Group Chat Workflow

```python
group_chat = GroupChat(
    agents = [
        user_agent,
        availability_agent,
        feature_agent,
        conflict_agent,
        booking_agent,
    ],
    messages = []
)
```

```python
chat_manager = GroupChatManager(
    groupchat = group_chat,
    name='SystemManager'
)
```

# Start Chat

```python
if __name__ == "__main__":
    user_agent.initiate_chat(
        chat_manager,
        message="Hi, I want to book a meeting room for 4 people with a projector on May 13 from 10 AM to 11 AM."
    )
```
